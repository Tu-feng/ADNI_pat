```python
import pandas as pd
import numpy as np
```

```python
recc = pd.read_csv('Data/Medical_History/RECCMEDS.csv')
```

```python
recc.columns
```

```python
f = open('candymed.txt')
lines = f.readlines()
```

```python
meds=lines
for i in range(len(meds)):
    meds[i]=meds[i][:-1]
```

```python
meds = meds[:-2]
```

```python
from collections import defaultdict
```

```python
patients = defaultdict(set)

for index in recc.index:
    data = recc.loc[index]
    for med in meds:
        if med.upper() in str(data['Medication']).upper() or med.upper() in str(data['If medication not found in search list, hand enter for approval']):
            patients[data['RID']].add(med)
```

```python
patients
```

```python
len(patients)
```

```python
n = len(patients)
out = [[0]*2 for _ in range(n)]
```

```python
for i,key in enumerate(patients.keys()):
    out[i][0] = key
    out[i][1] = tuple(patients[key])
```

```python
out
```

```python
c = pd.DataFrame(out,columns=['RID','Drug'])
```

```python
c
```

```python
c.to_csv('drugs.csv')
```
