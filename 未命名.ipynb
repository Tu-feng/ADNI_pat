```python
res = pd.DataFrame(columns = recc.columns)
```

```python
drugname = set()
for drug in drugs['Glyset']:
    if drug[-1]==' ':
        drug = drug[:-1]
    drugname.add(drug.upper())
```

```python
drugname
```

```python
for i in patrecc.index:
    if (str(patrecc.loc[i]['Medication']).upper() in drugname
        or str(patrecc.loc[i]['If medication not found in search list, hand enter for approval']).upper() in drugname):
            res.loc[i] = patrecc.loc[i]
        
```

```python
res.to_csv('Data/glymed_record.csv')
```

```python
f = open('candymed.txt')
lines = f.readlines()
```

```python
meds=lines
for i in range(len(meds)):
    meds[i]=meds[i][:-1]
```

```python
meds = meds[:-2]
```

```python
from collections import defaultdict
```

```python
patients = defaultdict(set)

for index in recc.index:
    data = recc.loc[index]
    for med in meds:
        if med.upper() in str(data['Medication']).upper() or med.upper() in str(data['If medication not found in search list, hand enter for approval']):
            patients[data['RID']].add(med)
```

```python
patients
```

```python
len(patients)
```

```python
n = len(patients)
out = [[0]*2 for _ in range(n)]
```

```python
for i,key in enumerate(patients.keys()):
    out[i][0] = key
    out[i][1] = tuple(patients[key])
```

```python
out
```

```python
c = pd.DataFrame(out,columns=['RID','Drug'])
```
